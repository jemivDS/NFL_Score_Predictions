```python
from sportsreference.nfl.boxscore import Boxscores
import pandas as pd
import numpy as np

## Create range of yeard from 1990 to 2018 and weeks from 1 to 17 (i.e. Regular Season)
years = list(np.arange(2012, 2019,1))
weeks = list(np.arange(1,18,1))



boxes = []
hometeam = []
awayteam = []
homescore = []
awayscore = []

for yr in years:
    for week in weeks:
        game = Boxscores(week, yr)
        
        games_sub = game.games
        key1 = list(games_sub.keys())[0]
        dict1 = games_sub[key1]
        for k in range(len(dict1)):
            dict2 = dict1[k]
            
            boxes.append(dict2['boxscore'])
            
            hometeam.append(dict2['home_name'])
            awayteam.append(dict2['away_name'])
            
            homescore.append(dict2['home_score'])
            awayscore.append(dict2['away_score'])
```

```python
boxes1 = pd.DataFrame({
    'box_key':boxes, 
    'home_team':hometeam, 
    'away_team':awayteam, 
    'home_score':homescore,
    'away_score':awayscore})        

keys = list(boxes1['box_key'])
```
